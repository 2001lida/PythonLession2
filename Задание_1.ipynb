```python
class Chocolate:
    def __init__(self, price: float, additive: str = None):
        if price <= 0:
            raise ValueError("Цена батончика должна быть больше нуля.")
        self.price = price
        self.additive = additive

    def show_my_chocolate(self):
      if self.additive is not None:
        print(f"Шоколадный батончик, добавка: {self.additive}")
      else:
          print("Обычный шоколадный батончик")
    def __str__(self):
        return f"Шоколадный батончик, {self.price} рублей"


# Демонстрация использования класса
if __name__ == "__main__":
    # 1. Батончик без добавки
    bar1 = Chocolate(price=1.99)
    print(bar1)
    bar1.show_my_chocolate()

    print("-" * 40)

    # 2. Батончик с добавкой
    bar2 = Chocolate(price=2.49, additive="орехи")
    print(bar2)
    bar2.show_my_chocolate()


    print("-" * 40)

    # 3. Ещё один батончик с другой добавкой
    bar3 = Chocolate(price=3.20, additive="карамель")
    print(bar3)
    bar3.show_my_chocolate()

    print("-" * 40)


    # 4. Батончик с отрицательной ценой (вызовет исключение)
    try:
        bar4 = Chocolate(price=-1.00)
    except ValueError as e:
        print(f"Ошибка: {e}")
```
